```python
transaction = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_transaction.csv')
identity = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_identity.csv')
```

```python
ignore = ['TransactionID', 'TransactionDT']
features = [col for col in transaction.columns if col not in ignore]

X = transaction[features]
```

```python
import mlflow
import pandas as pd
mlflow.set_tracking_uri("https://dagshub.com/gbend22/ml-homework-2.mlflow")
logged_model = 'runs:/ec6b3ca404254bc299b3ece920a6fcaa/XGBoost'

loaded_model = mlflow.sklearn.load_model(logged_model)
```

```python
predictions = loaded_model.predict_proba(X)[:, 1]

submission = pd.DataFrame({
    'TransactionID': transaction['TransactionID'],
    'isFraud': predictions
})
submission.to_csv('submission.csv', index=False)
submission
```
